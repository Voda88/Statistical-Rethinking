How to calculate the amount the probability of a certain sequence of binary events, when order is ignored?

$$ Pr( W \mid N,p) = \frac{N!}{W!(N-W)!} p^{W} (1-p)^{N-W}$$

Observing 6 head in 9 coin tosses, with p=0.5 using scipy:

```python
stats.binom.pmf(6, n=9, p=0.5)
```

$ Pr(W) $ and $ Pr(p) $ define prior predictive distribution

Bayesian estimates are a posterior distribution over parameters: $ Pr( parameters \mid data) $. Here: $ Pr(p \mid W,n) $

How to calculate posterior using grid approximation:

```python
def posterior_grid_approximation(grid_points = 5, success = 6, tosses = 9):
    #define grid
    p_grid = np.linspace(0,1,grid_points)
    
    #define prior
    prior = np.repeat(5,grid_points)
    
    #compute likelihood at each point in the grid
    likelihood = stats.binom.pmf(success, tosses, p_grid)
    
    #compute product of likelihood and prior
    unstd_posterior = likelihood*prior
    
    #standardize the posterior, so it sums to 1
    posterior = unstd_posterior/unstd_posterior.sum()
    return p_grid, posterior
    
```

```python
points = 20
w,n = 6,9
p_grid, posterior = posterior_grid_approximation(points, w,n)
```

```python
plt.plot(p_grid, posterior, 'o-', label = f"successes = {w}\ntosses = {n}")
plt.xlabel('probability of water', fontsize = 14)
plt.ylabel('posterior probability', fontsize = 14)
plt.title(f'{points}')
plt.legend(loc=0)
```

## Homework 2.1

what is the posterior distribution with 8 W and 15 tosses? Let's assume 50 grid points.

```python
w, n = 8, 15
points = 50
```

First, let's create our prior:

```python
prior = np.repeat(10,points)
```

Then lets create the array of grid points corresponding to different values of $Pr(p)$

```python
propabilities = np.linspace(0,1,points)
```

Let's create the probability distribution (binomial):

```python
distribution = stats.binom.pmf(w,n,propabilities)
```

How to create the posterior distribution?

```python
posterior = prior*distribution/(prior*distribution).sum()
```

```python
plt.plot(propabilities, posterior, 'o-')
```
